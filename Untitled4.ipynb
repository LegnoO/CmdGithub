```python
import numpy as np

n = 365 
a = 10
b = 40
data_uniform = np.random.uniform(a,b+1,n)
data_uniform.astype(int)
import matplotlib.pyplot as plt
plt.hist(data_uniform)
plt.title('Uniform Distribution')
plt.show()

p = (30-15)*(1/(b-a))
print('Xc xut doanh s hng ngy s gim trong khong t 15 n 30 =',p)

z = (b-20)*(1/(b-a))
print('Xc xut doanh s hng ngy ln hn 20 =',z)

e_x = (a+b)/2
print('Trung bnh =',e_x)
v_x = ((b-a)**2)/12
print('Phng sai =',v_x)
```

```python
from scipy.stats import binom
n = 12
p = 0.15
size = 1000
binom = binom.rvs(n=n,p=p,size=size)
binom

import seaborn as sns
ax = sns.distplot(binom,kde=False,color='blue',hist_kws={"linewidth":15,'alpha':1})
ax.set(xlabel='Binomial Distribution',ylabel='Frequency')

mean = n*p
print('mean =',mean)

q = 1-p
variance = n*p*q
print('variance',variance)
```

```python
from scipy.stats import norm
data_normal = norm.rvs(size = 1000,loc = 0,scale = 1)
import matplotlib.pyplot as plt
import seaborn as sns
ax = sns.distplot(data_normal,bins= 100,kde=True,color='blue',hist_kws={"linewidth":15,'alpha':1})
ax.set(xlabel='Normal Distribution',ylabel='Frequency')
E_X = data_normal.mean()
print('Mean',E_X)
V_X = data_normal.std() * data_normal.std()
print('Variance',V_X)
```
